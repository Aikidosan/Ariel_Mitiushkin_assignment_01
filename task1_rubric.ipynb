# Task 1 - Define Your Rubric

## 1. Criterion Definitions

### Fluency – natural, easy-to-read sentences

| Rating | Definition |
|--------|------------|
| good | All sentences are clear and easy to follow; no confusing phrasing. No sentence needs re-reading. Ideas flow logically with no abrupt jumps. |
| ok | Overall understandable, but 1–2 sentences feel slightly awkward or clunky. Reader may need to re-read at most one sentence. Flow is mostly logical with minor weak transitions. |
| bad | Multiple sentences are confusing, incomplete, or very awkward. Reader struggles to understand key information or must re-read several times. Text feels disjointed. |

### Grammar – spelling, punctuation, basic correctness

| Rating | Definition |
|--------|------------|
| good | 0–1 minor grammar, spelling, or punctuation errors in the whole description. No error changes the meaning or looks unprofessional. Consistent tense and correct sentence structure. |
| ok | 2–5 minor errors, but the text remains clearly understandable. No systematic error pattern. Errors do not significantly reduce professionalism. |
| bad | More than 5 errors, or any critical mistakes that make sentences hard to read. Systematic repeated errors. Description looks unprofessional or confusing. |

### Tone – friendly, credible sales voice

| Rating | Definition |
|--------|------------|
| good | Tone is clearly friendly and positive, but not exaggerated or spammy. Focuses on customer benefits and value. Sounds like a genuine sales description from a trusted store. |
| ok | Tone is mostly neutral or slightly friendly, but could be more engaging. Some benefit-focused language, but also some generic or flat sentences. No obviously inappropriate phrasing. |
| bad | Tone feels wrong for a sales page (rude, sarcastic, negative, or very robotic). Overly clickbait or unrealistic promises. Does not feel like a credible product description. |

### Length – 50–90 words

| Rating | Definition |
|--------|------------|
| good | Total word count between 50 and 90 words (inclusive). |
| ok | Total word count between 40–49 or 91–110 words (inclusive). Still long enough to understand the product. |
| bad | Fewer than 40 words or more than 110 words. Too short to give enough detail, or too long to be concise. |

### Grounding – sticks to the provided information

Inputs: product name, structured attributes, material, warranty.

| Rating | Definition |
|--------|------------|
| good | All concrete factual statements are supported by the given fields. No invented specs (e.g., battery life, dimensions, brand certifications) that are not in the data. Generic non-factual marketing phrases ("perfect for everyday use") are allowed. |
| ok | One minor inferred detail that is plausible but not strictly given (e.g., calling a cotton T-shirt "soft" when material is cotton). No major fabricated facts. |
| bad | Any wrong or clearly fabricated factual detail (e.g., changing material, adding a warranty length not given). More than one invented or risky assumption. Contradicts the input data. |

### Latency – average time per call

Let L = average latency in milliseconds across the evaluated calls.

| Rating | Definition |
|--------|------------|
| good | L ≤ 1500 ms per call |
| ok | 1500 < L ≤ 3000 ms per call |
| bad | L > 3000 ms per call |

### Cost – average price per call

Model: `google/gemma-2-9b-it-fast` on Nebius Token Factory.  
Pricing: **$0.03 / 1M input tokens**, **$0.09 / 1M output tokens** (base flavor).  
Let C = cost in USD per call (input_tokens × $0.03/1M + output_tokens × $0.09/1M).

| Rating | Definition |
|--------|------------|
| good | C ≤ $0.000030 per call (well within the model's baseline cost tier) |
| ok | $0.000030 < C ≤ $0.000100 per call (moderate, acceptable for production) |
| bad | C > $0.000100 per call (significantly above expected for this model) |

## 2. Pass / Fail Definition

### 2a. Cumulative Pass Bar

A description **passes** if ALL of the following are true:
- At least **4 criteria** are rated `good`
- No more than **2 criteria** are rated `ok`
- **0 criteria** are rated `bad`

Equivalently: *"Pass if rating is good or ok on every criterion, and at least four of them are good."*

### 2b. Go / No-Go Rules (Automatic Fail)

Regardless of the cumulative bar, a description is an **automatic fail** if any of these are true:
- **Grounding = bad** (any fabricated or conflicting factual information)
- **Length = bad** (fewer than 40 words or more than 110 words)
- **Fluency = bad** (text is hard to understand)
- **Grammar = bad** (many errors or clearly unprofessional)

Latency and Cost alone do not trigger automatic failure — they only affect pass/fail through the cumulative bar.

---

## 3. Formula (Compact)

```
Let ratings per criterion be in {good, ok, bad}.

A row FAILS automatically if:
    Grounding = bad  OR
    Length    = bad  OR
    Fluency   = bad  OR
    Grammar   = bad

Otherwise, let:
    G = number of criteria with rating = good
    O = number of criteria with rating = ok
    B = number of criteria with rating = bad

The row PASSES if and only if:
    G >= 4  AND
    O <= 2  AND
    B == 0
```
